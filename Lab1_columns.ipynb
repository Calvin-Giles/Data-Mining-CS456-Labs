## Creating a dataset

```python
import numpy as np
import pandas as pd
import random

#declares list for the data table
Student_ID_List = []
Score_List = []
Major_List = []
Grad_Year_List = []
Pass_Fail_List = []

#holds list of majors and years to randomly populate Major_List and Grade_Year_List
Majors = ['Computer Science', 'Mathmatics', 'Communications', 'History', 'Art', 'Biology', 'Astronomy', 'Business']
Years = [2025, 2026, 2027, 2028, 2029]

#for loops to randomly populate data 
for i in range (50):
    Student_ID_List.append(5000 + (i*10))

for i in range (50):
    Random_Float = random.random()
    Score_List.append(Random_Float)
    
for i in range (50):
    Major_List.append(random.choice(Majors))

for i in range (50):
    Grad_Year_List.append(random.choice(Years))

for i in range (50):
    if Score_List[i] < .70:
        Pass_Fail_List.append(False)
    else:
        Pass_Fail_List.append(True)
        
#creates the dictionary for the table        
Dictionary_Of_Data = {'Student ID': Student_ID_List, 'Score': Score_List, 'Student Major': Major_List, 'Graduation Year': Grad_Year_List, 'Passed': Pass_Fail_List}

#assignes the dataframe to manipulate the data
dframe = pd.DataFrame(Dictionary_Of_Data)
dframe
```

## Column challenge

```python
#add a new column that lists the letter grade from score
Letter_Grades = []

#populates the new column with the letter grade equivalent of score
for i in range (50):
    if Score_List[i] > .90:
        Letter_Grades.append('A')
    elif Score_List[i] > .80:
        Letter_Grades.append('B')
    elif Score_List[i] > .70:
        Letter_Grades.append('C')
    elif Score_List[i] > .60:
        Letter_Grades.append('D')
    else:
        Letter_Grades.append('F')
        
#adds the new Letter Grade column to the dataset        
dframe['Letter Grade'] = Letter_Grades
dframe
```
